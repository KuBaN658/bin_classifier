```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
X_train = np.loadtxt('data/X_train.csv', delimiter=',')
X_valid = np.loadtxt('data/X_valid.csv', delimiter=',')
y_train = np.loadtxt('data/y_train.csv', delimiter=',')
y_valid = np.loadtxt('data/y_valid.csv', delimiter=',')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC


lr = LogisticRegression()
svm = SVC(kernel='linear')

lr.fit(X_train, y_train)
svm.fit(X_train, y_train)
y_pred_lr = lr.predict(X_valid)
y_pred_svm = svm.predict(X_valid)
```

```python
from sklearn.metrics import f1_score


print(f'Вес одного из классов больше - '
      f'{y_valid[y_valid == 1].size / y_valid[y_valid == -1].size}')
```

```python
print(f"F1 мера линейной модели - {f1_score(y_valid, y_pred_lr)}")
print(f"F1 мера SVM - {f1_score(y_valid, y_pred_svm)}")
```

```python
import itertools

import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt

from mlxtend.plotting import plot_decision_regions

gs = gridspec.GridSpec(1, 2)

fig = plt.figure(figsize=(10,8))

labels = ['Logistic Regression', 'SVM']
for clf, lab, grd in zip([lr, svm],
                         labels,
                         itertools.product([0, 1], repeat=2)):

    ax = plt.subplot(gs[grd[0], grd[1]])
    fig = plot_decision_regions(X=np.array(X_train), y=np.array(y_train.astype(np.int32)), clf=clf, legend=2)
    plt.title(lab)

plt.show()
```
